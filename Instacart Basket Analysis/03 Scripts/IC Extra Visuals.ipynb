### 01. Extra IC Visuals

```python
# Creating a crosstab between 'spending_flag' and 'department'

spending_dept = pd.crosstab(df['department'], df['spending_flag'], dropna=False)
```

```python
# Creating a stacked bar chart between 'spending_flag' and 'department'
spending_dept_sbar = spending_dept.plot.barh(stacked=True,color=sns.color_palette('GnBu_r',3))
plt.title('Profile: Spending and Departments')
plt.legend(title = 'Spending Classification')
plt.ylabel('Departments')
plt.show()
```

```python
# Creating a crosstab between 'spending_flag' and 'region'

region_spending = pd.crosstab(df['region'], df['spending_flag'], dropna=False)
```

```python
# Creating a stacked bar chart between 'spending_flag' and 'region'
spending_region_sbar = region_spending.plot.bar(stacked=True,color=sns.color_palette('GnBu_r',3))
plt.title('Profile: Spending and Regions')
plt.legend(title = 'Spending Classification')
plt.ylabel('Region')
plt.show()
```

```python
# Creating a crosstab between loyalty and region
loyalty_region = pd.crosstab(df['region'],df['loyalty_flag'],dropna=False)
```

```python
# Creating a stacked bar chart between 'loyalty_flag' and 'region'
loyalty_region_sbar = loyalty_region.plot.bar(stacked=True,color=sns.color_palette('GnBu',3))
plt.title('Profile: Loyalty and Regions')
plt.legend(['New customer', 'Regular customer', 'Loyal customer'], title = 'Loyalty Classification')
plt.xlabel('Region')
plt.show()
```

```python
# Creating a crosstab between age and dependants
age_family = pd.crosstab(df['family_flag'], df['age_flag'], dropna=False)
```

```python
# Creating a stacked bar chart between age and family status
age_family_sbar = age_family.plot.bar(stacked=True,color=sns.color_palette('GnBu',3))
plt.title('Profile: Age and Dependants')
plt.legend(['Young adult', 'Middle-age', 'Senior'], title = 'Age Classification')
plt.xlabel('Family Status')
plt.show()
```

```python
# Price range profile

df['price_range_loc'].value_counts()
```

```python
price_range_pie = df['price_range_loc'].value_counts().plot.pie(title='Price Ranges',label='',autopct=('%1.f%%'))
```

```python
#Pie chart on loyalty

loyalty_pie = df['loyalty_flag'].value_counts().plot.pie(title='Customer Type', label='',autopct=('%1.f%%'))
```

```python
# Visualization customers by region

region_bar =df['region'].value_counts().plot.bar(color=sns.color_palette('GnBu_r',4))
plt.title('Customers by Region')
plt.xlabel('Region')
plt.show()
```

```python
# Creating a crosstab between age and dependants
loyalty_freq = pd.crosstab(df['days_since_prior_order'], df['loyalty_flag'], dropna=False)
```

```python
# Creating a stacked bar chart between age and family status
loyalty_freq_sbar = loyalty_freq.plot.barh(stacked=True,color=sns.color_palette('GnBu',3))
plt.title('Profile: Loyalty and Order Frequency')
plt.legend(title = 'Loyalty Classification')
plt.ylabel('Days Since Prior Order')
plt.show()
```

```python
#Creating a crosstab between hour and loyalty
loyalty_hour = pd.crosstab(df['order_hour_of_day'], df['loyalty_flag'], dropna=False)
```

```python
# Creating a stacked bar chart between age and family status
loyalty_hour_sbar = loyalty_hour.plot.barh(stacked=True,color=sns.color_palette('GnBu',3))
plt.title('Profile: Loyalty and Order Time of Day')
plt.legend(title = 'Loyalty Classification')
plt.ylabel('Hour of Day')
plt.show()
```

```python
#Creating a crosstab between department and loyalty
loyalty_dept = pd.crosstab(df['department'], df['loyalty_flag'], dropna=False)
```

```python
# Creating a stacked bar chart between loyalty and department
loyalty_dept_sbar = loyalty_dept.plot.barh(stacked=True,color=sns.color_palette('GnBu',3))
plt.title('Profile: Loyalty and Department')
plt.legend(title = 'Loyalty Classification')
plt.ylabel('Department')
plt.show()
```

```python
#Creating a crosstab between department and region
region_dept = pd.crosstab(df['department'], df['region'], dropna=False)
```

```python
# Creating a stacked bar chart between region and department
region_dept_sbar = region_dept.plot.barh(stacked=True,color=sns.color_palette('GnBu',4))
plt.title('Profile: Region and Department')
plt.legend(title = 'Region')
plt.ylabel('Department')
plt.show()
```

```python
#Creating a crosstab between department and family status
fam_dept = pd.crosstab(df['department'], df['family_flag'], dropna=False)
```

```python
# Creating a stacked bar chart between fam and department
region_dept_sbar = fam_dept.plot.barh(stacked=True,color=sns.color_palette('GnBu',2))
plt.title('Profile: Family Status and Department')
plt.legend(title = 'Family Status')
plt.ylabel('Department')
plt.show()
```
