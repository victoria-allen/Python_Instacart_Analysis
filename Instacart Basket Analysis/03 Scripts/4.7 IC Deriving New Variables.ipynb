```python
# Create a user-defined function to flag item prices

def price_label(row):

    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis =1)
```

#### 02. Create an if-statement with the loc() function

Create conditions for the loc() function. Use different cells for each condition.

```python
# Condition for high-range products

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# Condition for mid-range products

df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
# Condition for low-range products

df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# Value counts for price_range_loc

df['price_range_loc'].value_counts(dropna = False)
```

#### 03. If-statements with for-loops

Use a for-loop to determine the busiest and slowest days of the week

```python
# Check which day most orders take place

df_ords_prods_merged['orders_day_of_the_week'].value_counts(dropna = False)
```

##### Explanation
0, Saturday, is the busiest day of the week.

4, Wednesday, is the slowest day of the week.

```python
# Create a for-loop 
# First create a list

result = []

for value in df_ords_prods_merged["orders_day_of_the_week"]:
    if value == 0:
        result.append("Busiest day")
    elif value == 4:
        result.append("Least busy")
    else:
        result.append("Regularly busy")
```

```python
result
```

```python
# Create a new column to make 'result' more useful

df_ords_prods_merged['busiest_day'] = result
```

```python
# Check frequency of values for new column, 'busiest_day'

df_ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

##### Step 4: When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# Create a for-loop to find busiest hours of the day
# First check frequency of 'order_hour_of_day'

df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

##### Explanation
I'll choose the six busiest hours, those with over 2,500,000 orders, and the six least busy hours, those with less than 250,000 orders.

10 (10am), 11 (11am), 14 (2pm), 15(3pm), 13(1pm), 12(12pm) are the busiests hour of the day.

3 (3am), 4 (4am), 2 (2am), 5 (5am), 1 (1am), 0 (12am) are the slowest hours of the day.

```python
# For-loop for busiest hours of the day

result_ords = []

for value in df_ords_prods_merged["order_hour_of_day"]:
    if value == 10 or value == 11 or value == 14 or value == 15 or value == 13 or value == 12:
        result_ords.append("Most orders")
    elif value == 3 or value == 4 or value == 2 or value == 5 or value == 1 or value == 0:
        result_ords.append("Fewest orders")
    else:
        result_ords.append("Average orders")
```

```python
# Create a new column to make 'result_ords' more useful

df_ords_prods_merged['busiest_period_of_day'] = result_ords
```
